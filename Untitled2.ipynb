```python
import pandas as pd

# Baca file Excel
df = pd.read_excel("dataset_cnn_cleaned.xlsx")

# Ambil data teks dan label
text_data_file = df['Clean Narasi'].astype(str).tolist()
label_data_file = df['hoax'].astype(int).tolist()

# Data manual (20 data)
manual_data = [
    ("Vaksin COVID-19 bisa menyebabkan autisme", 0),
    ("Bumi berbentuk datar menurut penelitian", 0),
    ("Minum air garam bisa menyembuhkan COVID-19", 0),
    ("5G adalah penyebab utama virus corona", 0),
    ("Makan telur mentah bikin tubuh kebal", 0),
    ("Chip disuntikkan lewat vaksin", 0),
    ("Daun jambu bisa menyembuhkan diabetes", 0),
    ("NASA menemukan planet berpenghuni", 0),
    ("Minyak kayu putih bisa bunuh virus", 0),
    ("Buah naga bisa sembuhkan kanker", 0),
    ("Vaksin COVID-19 telah melalui uji klinis", 1),
    ("Pemerintah umumkan libur nasional 17 Agustus", 1),
    ("BMKG perkirakan hujan ringan di Jakarta", 1),
    ("WHO anjurkan vaksinasi booster", 1),
    ("Penelitian tunjukkan efektivitas masker", 1),
    ("Kemenkes buka program vaksin gratis", 1),
    ("ITB luncurkan mobil listrik karya mahasiswa", 1),
    ("Pesawat mendarat darurat di Soetta", 1),
    ("Pemerintah salurkan bantuan untuk UMKM", 1),
    ("Pertamina turunkan harga BBM", 1),
]

text_manual, label_manual = zip(*manual_data)

# Gabungkan semua data
all_texts = text_data_file + list(text_manual)
all_labels = label_data_file + list(label_manual)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
import numpy as np

# Tokenisasi
tokenizer = Tokenizer(num_words=5000, oov_token="<OOV>")
tokenizer.fit_on_texts(all_texts)
sequences = tokenizer.texts_to_sequences(all_texts)
padded_sequences = pad_sequences(sequences, maxlen=200, padding='post')

# Label
labels_array = np.array(all_labels)

# Split train-test
X_train, X_test, y_train, y_test = train_test_split(
    padded_sequences, labels_array, test_size=0.2, random_state=42, stratify=labels_array
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense

# Arsitektur CNN
model = Sequential([
    Embedding(input_dim=5000, output_dim=64),
    Conv1D(128, 5, activation='relu'),
    GlobalMaxPooling1D(),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Training
history = model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Evaluasi
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy:.2f}")
```

```python
predictions = (model.predict(X_test) > 0.5).astype("int32")
print("Prediksi:", predictions.flatten())
print("Asli:    ", y_test)
```
